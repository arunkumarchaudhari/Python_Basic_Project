```python
import numpy as np
import pandas as pd
import chart_studio.plotly as pl
import plotly.offline as po
import cufflinks as cf
```

```python
po.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
def createdata(data):
    if(data ==1):
        x = np.random.rand(100,5)
        df1 = pd.DataFrame(x, columns=['A','B','C','D','E'])
    elif(data==2):
        x = [0,0,0,0,0]
        r1 = [0,0,0,0,0]
        r2 = [0,0,0,0,0]
        r3 = [0,0,0,0,0]
        r4 = [0,0,0,0,0]
        print('enter the value for columns')
        i=0
        for i in [0,1,2,3,4]:
            x[i] =input()
            i=i+1
        print('enter the value for 1st row')
        i=0
        for i in [0,1,2,3,4]:
            r1[i] =int(input())
            i=i+1
        print('enter the value for 2nd row')
        i=0
        for i in [0,1,2,3,4]:
            r2[i] =int(input())
            i=i+1
        print('enter the value for 3rd row')
        i=0
        for i in [0,1,2,3,4]:
            r3[i] =int(input())
            i=i+1
        print('enter the value for 4th row')
        i=0
        for i in [0,1,2,3,4]:
            r4[i] =int(input())
            i=i+1
        df1 = pd.DataFrame([r1,r2,r3,r4], columns=x)
    
    elif(data==3):
        file = input('enter file name with extension:: ')
        x = pd.read_csv(file)
        df1 = pd.DataFrame(x)
    else:
        print('DataFrame creation failed. Please try again') 
    return df1
        
```

```python
def plotter(plot):
    if(plot==1):
        finalplot = df1.iplot(kind='scatter')
    elif(plot==2):
        finalplot = df1.iplot(kind='scatter', mode='markers', symbol='x', colorscale='paired')
    elif(plot==3):
        finalplot = df1.iplot(kind='bar')
    elif(plot==4):
        finalplot = df1.iplot(kind='hist')
    elif(plot==5):
        finalplot = df1.iplot(kind='box')
    elif(plot==6):
        finalplot = df1.iplot(kind='surface')
    else:
        print('enter valid option.')
    return finalplot
```

```python
def plotter2(plot):
    col = int(input('enter the number of columns want to plot(between 1-3)'))
    if(col==1):
        colm =input('enter column from dataframe head')
        if(plot==1):
            finalplot = df1[colm].iplot(kind='scatter')
        elif(plot==2):
            finalplot = df1[colm].iplot(kind='scatter', mode='markers',colorscale='paired', symbol='x')
        elif(plot==3):
            finalplot = df1[colm].iplot(kind='bar')
        elif(plot==4):
            finalplot = df1[colm].iplot(kind='hist')
        elif(plot==5):
            finalplot = df1[colm].iplot(kind='box')
        elif(plot==6 or plot==7):
            finalplot = print('Bubble plot and surface plot require more than one column')
        else:
            finalplot = print('enter valid input')
        
    elif(col==2):
        colm = input('enter 1st column from dataframe head')
        coln = input('enter 2nd columm from dataframe head')
        if(plot==1):
            finalplot = df1[[colm, coln]].iplot(kind='scatter')
        elif(plot==2):
            finalplot = df1[[colm,coln]].iplot(kind='scatter', mode='markers',colorscale='paired', symbol='x')
        elif(plot==3):
            finalplot = df1[[colm,coln]].iplot(kind='bar')
        elif(plot==4):
            finalplot = df1[[colm,coln]].iplot(kind='hist')
        elif(plot==5):
            finalplot = df1[[colm,coln]].iplot(kind='box')
        elif(plot==6):
            finalplot = df1[[colm,coln]].iplot(kind='surface')
        elif(plot==7):
            size = int(input('enter the size column for bubble plot'))
            finalplot = df1.iplot(kind='bubble', x=colm, y=coln, size=size)
        else:
            finalplot = print('enter valid input')
    elif(col==3):
        colm = input('enter 1st column from dataframe head')
        coln = input('enter 2nd columm from dataframe head')
        colo = input('enter 3rd column from dataframe head')
        if(plot==1):
            finalplot = df1[[colm, coln,colo]].iplot(kind='scatter')
        elif(plot==2):
            finalplot = df1[[colm,coln,colo]].iplot(kind='scatter', mode='markers',colorscale='paired', symbol='x')
        elif(plot==3):
            finalplot = df1[[colm,coln,colo]].iplot(kind='bar')
        elif(plot==4):
            finalplot = df1[[colm,coln,colo]].iplot(kind='hist')
        elif(plot==5):
            finalplot = df1[[colm,coln,colo]].iplot(kind='box')
        elif(plot==6):
            finalplot = df1[[colm,coln,colo]].iplot(kind='surface')
        elif(plot==7):
            size = int(input('enter the size column for bubble plot'))
            finalplot = df1.iplot(kind='bubble', x=colm, y=coln,z=colo, size=size)
        else:
            finalplot = print('enter valid input')
    else:
        finalplot = print("enter valid input")
    
    return finalplot
```

```python
def main(cat):
    if(cat==1):
        print('Select the type of plot')
        print('1. Line Plot')
        print('2. Scannter plot')
        print('3. Bar Plot')
        print('4. Histogram')
        print('5. Box plot')
        print('6. Surface plot')
        plot = int(input())
        output = plotter(plot)
        
    elif(cat==2):
        print('Select the type of plot')
        print('1. Line Plot')
        print('2. Scannter plot')
        print('3. Bar Plot')
        print('4. Histogram')
        print('5. Box plot')
        print('6. Surface plot')
        print('7. Bubble plot')
        plot = int(input())
        output = plotter2(plot)
    else:
        print('Enter valid input')
```
